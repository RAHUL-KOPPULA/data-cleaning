```python
original_shape = data.shape
null_summary_before = data.isnull().sum()
duplicates_count = data.duplicated().sum()
print(original_shape,'original shape')
print(null_summary_before,'null values')
print(duplicates_count,'duplicates')
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data
```

```python
data['director'].fillna('unknown',inplace=True)
data
```

```python
data.duplicated().sum()
data.duplicated()
```

```python
data.dropna()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data['type'] = data['type'].apply(lambda x: str(x).strip().lower().title() if pd.notnull(x) else x)
data[['type', 'title']].head()
```

```python
data['country'] = data['country'].apply(lambda x: str(x).strip() if pd.notnull(x) else x)
data[['type', 'country']].head()
```

```python
data.columns = [col.strip().lower().replace(" ", "_") for col in data.columns]
data
```

```python
data = data[pd.notnull(data['release_year'])]
data['release_year'] = data['release_year'].astype(int)
```

```python
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
data['date_added'] = data['date_added'].dt.strftime('%Y-%m-%d')
data = data[data['date_added'].notnull()]
data.notnull()
```

```python
print("✅ Summary of Data Cleaning\n")
print(f"Original number of rows: {original_shape[0]}")
print(f"Final number of rows: {data.shape[0]}")
print()
print("🧹 Null Values Handling:")
print(null_summary_before)
print("\nNull values after cleaning:")
print(data.isnull().sum())
print()
print(f"📑 Duplicate rows removed: {duplicates_count}")
print("\n🧾 Final Column Names:")
print(list(data.columns))
print("\n📊 Final Data Types:")
print(data.dtypes)
```
